## An Alternative Explanation for Factors

Fama와 French가 제안한 3요인 모델(1993)과 5요인 모델(2015)은 수천 편의 학술 논문에 영감을 주었으며, 다양한 팩터를 발견하는 템플릿 역할을 해왔다. $\text{FF93}$은 평균 주식 수익률의 횡단면을 Market, Size(size: 주가와 주식 수로 측정), Value(book-to-market 비율로 측정)라는 세 가지 요인의 선형 함수로 부분적으로 설명한다고 가정한다. $\text{FF15}$는 설명력을 개선하기 위해 수익성(profitability)과 투자(investment)라는 두 가지 품질 지표를 추가했다.

$\text{FF93}$과 $\text{FF15}$의 모델 명세에 대해 몇 가지 비판이 제기된다. 첫째, 모델을 선택하기 전 수행된 모든 실험을 보고하거나 조정하지 않았기 때문에 p-해킹이 발생했을 가능성이 있다(6.4.1.1절). 둘째, 인과 그래프가 아닌 설명력을 근거로 모델 명세를 정당화하여 가설 설정 오류(specification searching)를 초래할 수 있다(6.4.2.3절). 셋째, 인플레이션, GDP, 경기 순환 단계, 수익률 곡선의 기울기 등 거시경제 혼란 변수를 무시했다. $\text{FF93}$의 2.1.2절에서는 경기 순환이 크기에 미치는 혼란 효과를 명시적으로 언급하고 있지만, 모델에서는 해당 혼란 변수를 배제하고 있다. 이는 누락된 혼란 변수의 문제를 시사한다(6.4.2.1절). 넷째, Momentum과 Value Factor 간의 상호작용이 잘 문서화되어 있다(Barroso and Santa-Clara, 2015). 이 상호작용은 Momentum과 Value 간의 혼란 관계로 설명될 수 있으며, Momentum이 또 다른 누락된 Confounder일 가능성을 제기한다. 다섯째, 제시된 관측값을 설명하는 인과 메커니즘을 제공하지 않음으로써 모델의 인과적 내용을 부정하고, 중재 분석(mediation analysis)과 반증(falsification) 노력을 방해한다(6.3절).

\begin{equation}
\text{FF93} : R_{i} - R_{f} = \alpha + \beta_1 (R_{m} - R_{f}) + \beta_2 \, \text{SMB} + \beta_3 \, \text{HML} + \varepsilon 
\end{equation}

\begin{equation}
\text{FF15} : R_{i} - R_{f} = \alpha + \beta_1 (R_{m} - R_{f}) + \beta_2 \, \text{SMB} + \beta_3 \, \text{HML} + \beta_4 \, \text{RMW} + \beta_5 \, \text{CMA} + \varepsilon 
\end{equation}


Carhart (1997) (이하 $\text{C97}$)은 $\text{FF93}$ 모델에 모멘텀을 네 번째 요인으로 추가했다. 

\begin{equation}
\text{C97} : R_{i} - R_{f} = \alpha + \beta_1 (R_{m} - R_{f}) + \beta_2 \, \text{SMB} + \beta_3 \, \text{HML} + \beta_4 \, \text{MOM} + \varepsilon 
\end{equation}

그러나 이 확장의 근거는 모멘텀 통제가 가치 요인의 인과 효과 추정의 혼란을 제거했다기보다는, 네 요인 모델이 더 높은 설명력을 제공한다는 연관성 기반 주장이었다. 이는 $\text{FF915}$가 품질 요인 두 개를 추가하며 사용한 자기모순적 논리와 동일하다(6.4.2.3절 참조). 7.3절에서 설명한 바와 같이, 올바른 모델 명세가 동일한 데이터셋에서 잘못된 모델보다 더 낮은 설명력을 제공할 수 있다. $\text{FF93}$의 설명력을 개선했다고 해서 $\text{C97}$의 모델이 더 잘 명세되거나 편향이 줄어든 것은 아니다. 또한, Momentum(혼란 제거 변수)과 Value(혼란된 변수)는 높은 상관관계를 가지며, 이는 혼란 편향을 다중공선성 문제로 대체할 뿐이다. 부분 상관법 대신 backdoor adjustment 이나 do-calculus 조정을 적용하면 공분산 행렬의 반전으로 인한 다중공선성을 피할 수 있다.

$\text{FF93}$, $\text{FF15}$, $\text{97}$ 모델의 설정에 대해 여섯 번째로 제기될 수 있는 비판은 다음과 같다. (1) 주식 수익률과 기업의 크기가 독립적 변수라고 가정하고, (2) 이 두 변수가 모두 기업의 **book-to-market ratio(collider)** 에 영향을 미친다고 할 때, book-to-market ratio를 조건부화하면 두 독립 변수 간에 부정적인 비인과적 연관이 유도된다(6.4.2.2절과 7.2절 참조). 즉, $\text{FF93}$, $\text{FF15}$, $\text{97}$의 모델 명세에 book-to-market ratio를 추가함으로써 주식 수익률과 크기 간에 거짓 부정적 상관관계가 발생하여 크기 요인을 잘못된 발견으로 만들 수 있다.

7절의 수치 실험은 일반적인 인과 구조를 통해 $\text{FF93}$, $\text{FF15}$, $\text{97}$의 발견이 B유형의 가짜 관계임을 설명한다. 아래의 그림은 혼란 변수와 콜라이더로 인해 $\text{FF93}$, $\text{FF15}$, $\text{97}$의 추정치에 편향이 발생하는 인과 그래프의 예시를 제공한다. 이 그래프가 정확하지 않을 수 있지만, 이 그래프가 잘못되었음을 증명할 책임은 투자 요인의 존재를 주장하는 저자들에게 있다. 이러한 문제를 해결하려면, 저자들은 모델의 인과적 내용을 명확히 밝히고, 가정된 인과 메커니즘을 선언하며, 누락된 혼란 변수를 통제하고, 선택된 설명 변수가 collider가 아님을 정당화해야 한다.

```python
import graphviz as gr

chain = gr.Digraph(
    'G',
    node_attr = {'shape': 'ellipse'}, 
    graph_attr = {
        'rankdir': 'LR',
        'label': 'Example of a causal graph under which the findings \n in FF93, FF15, and C97 are biased'
    }
)

chain.edge('Market', 'Ret', label = '1', color = 'red')
chain.edge('MOM', 'Ret', label = '5', color = 'red')
chain.edge('MOM', 'HML', label = '4', color = 'red')
chain.edge('SML', 'HML', label = '3', color = 'red')
chain.edge('Ret', 'HML', label = '2', color = 'red')

chain.node(
    'HML', style = 'filled', 
    fillcolor = 'lightgray', 
    shape = 'ellipse', 
    penwidth = '0'
)

with chain.subgraph() as s:
    s.attr(rank = 'same') 
    s.node('Ret')
    s.node('HML')

chain
```

$\text{FF93}$, $\text{FF15}$, 그리고 $\text{97}$이 예측 모델을 제안했다면 팩터 계수의 편향된 추정치는 큰 문제가 되지 않았을 것이다. 예측이 여전히 수익성 있는 투자 전략으로 이어질 수 있기 때문이다. 그러나 6.1절에서 설명하듯, 팩터 모델의 목적은 특정 위험 특성에 노출된 포트폴리오를 구축하는 것이다. 이러한 이유로 Value Factor 펀드는 잔차 대신 book-to-market 비율로 주식을 순위 매겨 포트폴리오를 구성한다.

정리하면, $\text{FF93}$, $\text{FF15}$, $\text{97}$의 발견은 p-hacking에 의한 A유형의 오류이거나, 혼란 변수를 과소 통제, 매개 변수를 과도 통제하고 명세 검색과 중재 분석 누락으로 인한 B유형의 오류일 가능성이 크다. 그러나 이것은 이 세 논문만의 문제가 아니다. 사실, 이 세 논문은 여전히 연관성을 다루는 금융 계량경제 연구의 모범 사례로서 강한 영향을 미치고 있다. 하지만 이 분야의 다른 연구들도 비슷한 방법론적 오류를 공유하며, 그들이 제시한 팩터 또한 가짜일 가능성이 높다.

```python
import pandas_datareader as pdr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

ff15_model = pdr.get_data_famafrench(
    'F-F_Research_Data_5_Factors_2x3_daily',
    start = '2004-02-02'
)
```

```python
c97_model = pdr.get_data_famafrench(
    'F-F_Momentum_Factor_daily',
    start = '2004-02-02'
)
```

```python
data = pd.concat(
    [ff15_model[0], c97_model[0]],
    axis = 1
)
```

```python
import yfinance as yf

price = yf.download(
    'VTV',
    start = '2004-01-30',
    end = '2024-09-01',
    progress = False
)
```

```python
ret = price['Adj Close'].pct_change().dropna()
ret.name = 'return'
```

```python
data = data * 0.01
```

```python
excess_return = ret - data['RF']
excess_return.name = 'excess_return'
data.drop(['RF'], axis = 1, inplace = True)
```

```python
data.columns = ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'MOM']
```

### Assumption : HML is Confounder

$$
\begin{align}
\text{\widehat{HML}}_t = & \alpha + \beta_1 \text{MOM}_t + \beta_2 \text{SMB}_t + \varepsilon_{\text{HML}} \\
R_t - R_f = & \beta_0 + \beta_1 (R_m - R_f) + \beta_2 \text{MOM}_t + \gamma \varepsilon_{\text{HML}} + \varepsilon
\end{align}
$$  

```python
import statsmodels.api as sm

T = sm.add_constant(
    data[['MOM','SMB']]
)
y_t = data['HML']
```

```python
ols0 = sm.OLS(y_t, T).fit()
```

```python
print(ols0.summary()) # true positive based causal graph
```

```python
hml_hat = ols0.resid
hml_hat.name = 'HML_hat'
```

```python
X = sm.add_constant(
    pd.concat(
        [data[['Mkt-RF','MOM']], hml_hat],
        axis = 1
    )
)
```

```python
ols1 = sm.OLS(excess_return, X).fit()
```

```python
print(ols1.summary())
```
