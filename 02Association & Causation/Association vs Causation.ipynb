### Association vs Causation

통계학과, 계량경제학을 공부하는 모든 학생은 '연관성이 인과관계를 의미하지 않는다'는 것을 배운다. 이 명제는 겉보기에는 사실이지만, 왜 연관성이 존재하는지, 그리고 그것이 인과관계와 어떤 관련이 있는지를 설명하지 않는다. 

두 이산 확률 변수 X와 Y가 통계적으로 독립적일 때는 $P[X = x, Y = y] = P[X = x]P[Y = y]$를 모든 $x, y$에 대해 만족한다. 여기서 $P[\cdot]$는 괄호 안에 설명된 사건의 확률이다. 반대로, 두 이산 확률 변수 $X$와 $Y$가 어떤 $(x, y)$에 대해 아래의 조건을 만족할 때, 통계적으로 연관되었다고 (또는 공변한다고) 말한다

$$P[X = x, Y = y] ≠ P[X = x]P[Y = y]$$

조건부 확률 표현 

$$P[Y = y|X = x] = P[X = x, Y = y] / P[X = x]$$ 

은 $X = x$인 전체 부분집합에서 $Y = y$일 확률을 나타낸다. 두 변수가 연관되어 있을 때, 하나의 값을 관찰하는 것은 다른 하나의 값에 대한 정보를 전달한다

$$P[Y = y|X = x] ≠ P[Y = y]$$

또는 동등하게 아래를 의미한다

$$P[X = x|Y = y] ≠ P[X = x]$$

월별 익사 사건 수와 아이스크림 판매량의 사례를 살펴보자. 월별 익사 사건 수$(Y)$와 아이스크림 판매량$(X)$은 강하게 연관되어 있다. 왜냐하면 그 달에 $x$개의 아이스크림이 팔린 것을 관찰했을 때, 그 달에 $y$명이 익사할 확률이 그 달에 $y$명이 익사할 무조건적 확률과 다르기 때문이다. 그러나, $P[Y = y|X = x] ≠ P[Y = y]$라는 표현은 아이스크림 판매가 익사 사고를 일으킨다고 말해주지 않는다. 그 질문에 답하기 위해서는 조건부 확률보다 더 미묘한 개념인 **'intervention'** 을 도입해야 한다

데이터 생성 과정은 관찰된 데이터를 생성하는 물리적 과정을 말하며, 이 과정은 구조 방정식의 체계에 의해 특징지어진다. 그 체계 내에서, 변수 $X$가 변수 $Y$를 유발한다고 할 때 $Y$는 $X$의 함수라고 할 수 있다. $X$가 $Y$를 유발하는 구조 방정식을 **인과 메커니즘**이라고 한다. 불행히도, 관찰에 인과성을 가지는 데이터 생성 과정은 거의 알려져 있지 않다. 대신, 연구자들은 시스템의 인과 구조를 추론하기 위해 관찰의 샘플에 기반한 확률에 의존해야 한다

확률적으로, 변수 $X$가 변수 $Y$를 불러 일으킨다고 할 때 $X$의 값을 $x$로 설정하는 것은 $Y$가 $y$ 값을 취할 가능성을 증가시키는 것을 의미한다. 계량경제학은 이러한 intervention을 표현하는 언어가 부족하다. 즉, $X$의 값을 설정하는 것이 바로 intervention을 표현하는 것의 전부이다(Chen and Pearl 2013)

$X = x$에 의한 조건화와 $X = x$ 값 설정 사이의 혼란을 피하기 위해, Pearl(1995)은 do 연산자, $do[X = x]$,를 도입했다. 이는 $X$의 값을 $x$로 설정하는 개입을 나타낸다. 이 새로운 표기법을 사용하여, 인과관계는 다음과 같이 공식적으로 정의될 수 있다 

<center>

$X = x$가 $Y = y$를 유발한다는 것은 $P[Y = y| \mbox{do}[X = x]] > P[Y = y]$일 때, 그리고 그럴 때에만 가능하다

</center>

예를 들어, 아이스크림 판매량을 $x$로 설정하는 것이 어떤 $(x, y)$ 쌍에 대해서도 조건 없는 확률보다 $y$명이 익사할 가능성을 더 높게 만들지 않으므로, 아이스크림 판매량은 익사 사고의 원인이 아니다. 반면에, 담배를 피우는 것은 폐암의 원인이다. 왜냐하면 담배를 피우는 수준이 $x$로 설정된(개입을 통해) 집단에서 $y$명이 폐암에 걸릴 확률이, 어떤 $(x, y)$ 쌍에 대해, $y$명이 폐암에 걸릴 무조건적 확률보다 높기 때문이다

변수 $X$와 $Y$는 추가 변수들을 포함하는 더 복잡한 시스템의 일부일 수 있다. 시스템의 인과 구조는 방향성 비순환 그래프(directed acyclic graph), 또는 인과 그래프로 표현될 수 있다. 인과 그래프는 데이터 생성 과정을 완전히 특징지어 주지는 않지만, 인과 효과를 추정하는 데 필수적인 위상학적 정보를 전달한다. 인과 그래프는 시스템에 관련된 변수들을 선언하고, 어떤 변수들이 서로에게 영향을 미치며, 인과성의 방향을 나타내는 그래프이다.(Pearl 2009, p. 12) 

do-작업을 인과 그래프에서 $X$를 향한 모든 화살표를 제거하는 행동으로 시각화하는 데 도움을 준다. 이는 $X = x$로 설정함으로써 $Y$에 대한 전체 효과를 귀속시킬 수 있게 한다. 이것이 다른 조건이 동일하다는(ceteris paribus) 가정의 의미이며, 이는 경제학자들에게 매우 중요하다

그림 1의 인과 그래프는 $Z$가 $X$를 유발하고, $Z$가 $Y$를 유발한다고 말해준다 인과 추론의 언어로, $Z$는 혼동 변수이다. 왜냐하면 이 변수는 $X$와 $Y$ 사이의 연관성을 도입하지만, $X$와 $Y$ 사이에 화살표가 없기 때문이다. 

이 때문에, 이런 유형의 연관성은 비인과적(noncausal)이라고 표시된다. 이전 예시로, 날씨($Z$)는 아이스크림 판매($X$)와 수영하는 사람의 수, 그리고 익사($Y$)에 영향을 준다. 아이스크림 판매를 설정하는 개입은 화살표(1)를 없앤다. 왜냐하면 연구자가 $X$를 완전히 통제하기 때문이다($X$는 더 이상 $Z$의 함수가 아니라고 보면 된다). 모든 다른 것들은 동일하게 유지된다(ceteris paribus). 그리고 $X$가 $Y$를 유발하지 않기 때문에, $X$를 $x$로 설정하는 것(예: 아이스크림 판매 금지, $X = 0$)은 $Y$가 $y$일 확률에 영향을 주지 않는다. 나중에 보겠지만, 비인과적 연관성은 혼동 변수를 포함하지 않는 다양한 추가 이유로 발생할 수 있다

이 서술에서 다섯 가지 결론을 도출할 수 있다

첫째, 인과성은 통계 외적인(관찰을 넘어서는 의미에서) 개념으로, 메커니즘과 개입에 연결되어 있으며 연관성 개념과는 다르다. 결과적으로, 연구자들은 조건부 확률의 연관성 언어로 인과 시스템을 설명할 수 없다. do-연산자를 사용하지 않는 것은 계량 경제학 등에서 연관성과 인과성 진술 사이의 혼란을 초래했었다

둘째, 연관성은 인과성을 의미하지 않지만, 인과성은 연관성을 의미한다. 왜냐하면 개입을 통해 $X = x$로 설정하는 것은 결과 $Y = y$와 연관되어 있기 때문이다 

셋째, 연관성과 달리 인과성은 방향성이 있다. 이것은 인과 그래프의 화살표로 표현된다. "$X$가 $Y$를 유발한다"는 진술은 $P(Y = y | do(X = x)) > P(Y = y)$를 의미하지만, $P(X = x | do(Y = y)) > P(X = x)$를 의미하지는 않는다

넷째, 연관성과 달리 인과성은 순차적이다. "$X$가 $Y$를 유발한다"는 것은 $X$의 값이 먼저 설정되고, 그 후에 $Y$가 적응한다는 것을 의미한다
 
다섯째, ceteris paribus 가정은 개입(do-연산)을 시뮬레이션하며, 그 함의는 인과 그래프의 지식만으로 이해될 수 있다. 인과 그래프는 개입에 의해 동일하게 유지되는 "다른 것들"이 무엇인지 보여준다
